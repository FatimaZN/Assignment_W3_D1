## Task 1 — BMI Category Checker

Create two variables `weight` (in kg) and `height` (in meters), for example `weight = 70` and `height = 1.75`.

1. Calculate BMI using the formula: `bmi = weight / (height ** 2)`
2. Print the BMI value
3. Print the category based on BMI:
   - Below 18.5 → `"Underweight"`
   - 18.5 to 24.9 → `"Normal"`
   - 25 to 29.9 → `"Overweight"`
   - 30 and above → `"Obese"`

```python
# Your code here
weight = 70 
height = 1.75 
bmi = weight/(height **2)
print(bmi)
if bmi < 18.5:
    print("Underweight")
elif 18.5 <= bmi <= 24.9: 
    print("Normal")
elif 25 <= bmi <= 29.9:
    print("Overweight")
else:
    print("obese")
```

## Task 2 — Simple Calculator with Decision

Create three variables: `a = 10`, `b = 0`, and `operation = "divide"`.

Based on `operation`, perform and print the result:
- `"add"` → a + b
- `"subtract"` → a - b
- `"multiply"` → a * b
- `"divide"` → a / b **but** first check: if `b == 0`, print `"Cannot divide by zero!"` instead
- anything else → print `"Unknown operation"`

```python
a = 10
b = 0
operation = "divide"

if operation == "add":
    print(a + b)
elif operation == "subtract":
    print(a - b)
elif operation == "multiply":
    print(a * b)
elif operation == "divide":
    if b == 0:
        print("Cannot divide by zero!")
    else:
        print(a / b)
else:
    print("Unknown operation")
```

## Task 3 — Password Strength Checker (stretch)

Given a variable `password = "Ali123"`, check and print whether it is strong.

A password is **strong** if BOTH:
- its length is at least 6 characters (`len(password) >= 6`), AND
- it is not one of the common weak passwords in this list: `["123456", "password", "12345678", "qwerty"]`

Print `"Strong password ✅"` or `"Weak password ❌"` accordingly.

*Hint: combine `len(...)` with `and` and `not ... in ...`.*

```python
# Your code here
password = "Ali123"
weak_password = ["123456", "password", "12345678", "qwerty"]

if (len(password) >= 6) and password not in weak_password: 
    print("Strong password")
else: 
    print("Weak password")
```
